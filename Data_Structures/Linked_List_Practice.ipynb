```python
class Node:
    def __init__(self,data=None,next=None):
        self.data = data
        self.next = next
        
```

```python
class LinkedList:
    def __init__(self):
        self.head = None
    
    def print_list(self):
        if self.head is None:
            print("Linked List is Empty")
            return
        itr = self.head
        result = ""
        while itr:
            result += str(itr.data) + "-->"
            itr = itr.next
        print(result)
            
    def insert_at_start(self,value):
        if self.head is None:
            self.head = Node(value)
        else:
            new_node = Node(value,self.head)
            self.head = new_node
    
    def insert_at_end(self,value):
        if self.head is None:
            self.head = Node(value)
        
        itr = self.head
        while itr.next:
            itr = itr.next
        itr.next = Node(value)
        
    def get_size(self):
        size = 0
        itr = self.head
        while itr:
            size +=1
            itr = itr.next
        return size
        
    
    def insert_at_index(self,index,value):
        if index < 0 or index > self.get_size():
            raise Exception("Out of index")
        if index == 0:
            self.insert_at_start(value)
        itr = self.head
        count = 0
        while itr:
            if count == index-1 :
                new_node = Node(value,itr.next)
                itr.next = new_node
            count += 1
            itr = itr.next
        
    def remove_at_index(self,index):
        if index < 0 or index > self.get_size():
            raise Exception("Out of index")
        if index == 0:
            self.head = self.head.next
        itr = self.head
        count = 0
        while itr:
            if count == index-1:
                itr.next = itr.next.next
            count+=1
            itr = itr.next
            
    def insert_after_value(self,after,value):
        itr = self.head
        while itr:
            if itr.data == after:
                new_node = Node(value,itr.next.next)
                itr.next = new_node
            itr = itr.next
                
    def remove_by_value(self,value):
        itr = self.head
        while itr.next:
            if itr.next.data == value:
                itr.next = itr.next.next
            itr = itr.next
```

```python
ll = LinkedList()
ll.print_list()
ll.insert_at_start(10)
ll.print_list()
ll.insert_at_start(100)
ll.insert_at_start(30)
ll.insert_at_start(50)
ll.print_list()
ll.insert_at_end(32)
ll.print_list()
print(ll.get_size())
ll.insert_at_index(2,500)
ll.insert_at_index(0,500)
ll.print_list()
ll.remove_at_index(2)
ll.print_list()
ll.insert_after_value(50,80)
ll.print_list()
ll.remove_by_value(80)
ll.print_list()
```
